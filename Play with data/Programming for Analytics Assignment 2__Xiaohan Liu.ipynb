### PART - 2: Messy Data Files.

#### Find the data file "hotels.txt" on Canvas and compute the following questions. 

##### Question 1: Count the number of observations.

```python
# YOUR CODE HERE
df1=pd.read_csv('C:/Users/xiaoh/Desktop/SMU Coursework/Mod B/Programming/Homework/Homework2/hotels.txt')
print('Number of Observation is', df1.shape[0])
```

##### Question 2: How many hotel types are there and how many of each?

```python
# YOUR CODE HERE
df1.groupby(["hotel_type"]).size().reset_index(name='counts')
```

##### Question 3: Count the values (frequency) of each of the categorical variables.

```python
# YOUR CODE HERE
catvals = list(df1.select_dtypes(include=['O']).columns) 
freq = [df1[c].value_counts() for c in catvals] 
print(freq)
```

##### Question 4: Which variables should be treated as text but the raw data presents them as numbers?

```python
# YOUR CODE HERE
df1.dtypes
for col in ["hotel_code", "location_code"]:
    df1[col] = df1[col].astype(str)
df1.dtypes
```

##### Question 5: Count the number of missing values in each column.

```python
# YOUR CODE HERE
df1.isnull().sum()
```

##### Question 6: Calculate the mean, standard deviation, skewness and kurtosis for each of the numeric variables.

```python
# YOUR CODE HERE

for x in list(df1.columns):
    if df1[x].dtypes == 'float64' or df1[x].dtypes =='int64':
        print(df1[x].agg(['mean','std','skew','kurtosis']))
```

##### Question 7: Count the number of outliers in each numeric column.

```python
# YOUR CODE HERE
low = 0.0015
high = 0.9985
std_df = df1.quantile([low,high])

for x in df1.columns:
    if df1[x].dtypes == "float64" or df1[x].dtypes =="int64":
        n = df1[x][df1[x] < std_df[x].iloc[0]].count() + df1[x][df1[x] > std_df[x].iloc[1]].count()
        print("variable", x, "has", n, "outliers")
```

##### Question 8: Change the outliers to missing.

```python
# YOUR CODE HERE
import numpy as np 
newdf1 = df1[std_df.columns]

stdlow_df = df1.quantile(low)
stdhigh_df = df1.quantile(high)
df1.fillna(-999999)

outlierlow= (newdf1 < stdlow_df)
outlierhigh = (newdf1 > stdhigh_df)

newdf1[outlierlow] = np.nan
newdf1[outlierhigh] = np.nan

for x in newdf1.columns:
    df1[x] = newdf1[x]

pd.options.mode.chained_assignment = None
df1.describe()
```

##### Question 9: Impute the missing values.

```python
# YOUR CODE HERE
replace = df1
for x in list(df1.columns):
    if replace[x].dtypes == 'int64' or replace[x].dtypes == 'float64':
        replace[x].fillna(value = replace[x].mean(), inplace = True)
```

##### Question 10: Compute the mean, standard deviation, skewness and kurtosis for each of the numeric variables after imputation.

```python
# YOUR CODE HERE
for x in list(df1.columns):
    if df1[x].dtypes == 'float64' or df1[x].dtypes =='int64':
        print(df1[x].agg(['mean','std','skew','kurtosis']))
```
