```python
# YOUR CODE HERE
import pandas as pd

# get the data
hotels = pd.read_csv("C:/Users/jhb_f/Desktop/Fall 2017 B/Programming/data/hotels.txt")
print(len(hotels))
```

```python
# YOUR CODE HERE
hotels.dtypes
hotels[['hotel_code', 'location_code']] = hotels[['hotel_code', 'location_code']].astype(str)
hotels.dtypes
```

```python
# YOUR CODE HERE
import numpy as np

#select only numeric colunms
hotelsnum = hotels.select_dtypes(exclude=['object'])

#assign outlers 
hotels_outliers = ((hotelsnum < (hotelsnum.mean() - 3 * hotelsnum.std())) 
                   | (hotelsnum > (hotelsnum.mean() + 3 * hotelsnum.std())))

#replace outliers to nan
hotelsnum1 = hotelsnum[hotelsnum.columns]
hotelsnum1[hotels_outliers] = np.nan

print(hotelsnum1.isnull().sum())
print(hotelsnum1.describe())
```

```python
print(hotelsnum1.isnull().sum())
print(hotels.isnull().sum())
```

```python
# YOUR CODE HERE
hotelsnum2 = hotelsnum1.fillna(hotelsnum1.mean())

print(hotelsnum2.isnull().sum())
print(hotelsnum2.describe())
```

```python
# YOUR CODE HERE
hotelsnum2.agg(['mean','std','skew','kurtosis'])
```

```python
hotels.update(hotelsnum2)
```

```python
print(hotels.isnull().sum())
```
