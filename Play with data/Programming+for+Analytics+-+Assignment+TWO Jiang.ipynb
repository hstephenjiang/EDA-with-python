#### Find the data file "hotels.txt" on Canvas and compute the following questions. 

##### Question 1: Count the number of observations.

```python
# YOUR CODE HERE
import pandas as pd

# get the data
hotels = pd.read_csv("C:/Users/jhb_f/Desktop/Fall 2017 B/Programming/data/hotels.txt")
print(len(hotels))
```

##### Question 2: How many hotel types are there and how many of each?

```python
# YOUR CODE HERE
hoteltypes = hotels.groupby('hotel_type').hotel_type.agg(['count'])
print(hoteltypes)
```

##### Question 3: Count the values (frequency) of each of the categorical variables.

```python
# YOUR CODE HERE
# assign all cat variable to a list
hotels_cat = list(hotels.select_dtypes(include=['object']).columns) 

#count the frequency
hotels_cat_freq = [hotels[cat].value_counts() for cat in hotels_cat] 
print(hotels_cat_freq)
            
```

##### Question 7: Count the number of outliers in each numeric column.

```python
# YOUR CODE HERE
#select only numeric colunms
hotelsnum = hotels.select_dtypes(exclude=['object'])

#count outliers out of +/-3 std
((hotelsnum < (hotelsnum.mean() - 3 * hotelsnum.std())) 
                   | (hotelsnum > (hotelsnum.mean() + 3 * hotelsnum.std()))).sum()
```

##### Question 8: Change the outliers to missing.

```python
# YOUR CODE HERE
import numpy as np

#assign outlers 
hotels_outliers = ((hotelsnum < (hotelsnum.mean() - 3 * hotelsnum.std())) 
                   | (hotelsnum > (hotelsnum.mean() + 3 * hotelsnum.std())))

#replace outliers to nan
hotelsnum1 = hotelsnum[hotelsnum.columns]
hotelsnum1[hotels_outliers] = np.nan

print(hotelsnum1.isnull().sum())
print(hotelsnum1.describe())
```

##### Question 9: Impute the missing values.

```python
# YOUR CODE HERE
hotelsnum2 = hotelsnum1.fillna(hotelsnum1.mean())

print(hotelsnum2.isnull().sum())
print(hotelsnum2.describe())
```

##### Question 10: Compute the mean, standard deviation, skewness and kurtosis for each of the numeric variables after imputation.

```python
# YOUR CODE HERE
hotelsnum2.agg(['mean','std','skew','kurtosis'])
```
