```python
def junk(j, k) :
    if j < 5 :
        z = j + k
    else :
        z = j * k
    return z 
```

```python
x = 5
y = 2

g = junk(x, y)
print(g)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 10]
```
